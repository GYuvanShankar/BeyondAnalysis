```python
import tensorflow as tf
import numpy as np
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
```

```python
train=pd.read_csv('train.csv')
test=pd.read_csv('test.csv')
```

```python
enc = OneHotEncoder(handle_unknown='ignore')
enc_df = pd.DataFrame(enc.fit_transform(train[['CATEGORY_1','CATEGORY_2']]).toarray())
a=pd.DataFrame(enc.transform(test[['CATEGORY_1','CATEGORY_2']]).toarray())
train=train.join(enc_df)
test=test.join(a)
```

```python
train=train.drop(columns=['CATEGORY_1','CATEGORY_2'])
test=test.drop(columns=['CATEGORY_1','CATEGORY_2'])
```

```python
seq_max_length=int(test.max(axis=0)['SEQUENCE_NO'])
```

```python
index=pd.unique(train['UNIQUE_IDENTIFIER'])
```

```python
count_train=int(len(index))
```

```python
train_y=train[['Y1','Y2']]
train_x=train.drop(columns=['Y1','Y2'])
```

```python
X_train=np.zeros((count_train, seq_max_length, 38))
```

```python
Y_train=np.zeros((count_train, 2))
seq_length_train=np.zeros((count_train))
```

```python
j=0
for i in index:
    df=train_x[train_x['UNIQUE_IDENTIFIER']==i]
    seq_length_train[j]=np.max(df['SEQUENCE_NO'].to_numpy())
    df=df.drop(columns=['UNIQUE_IDENTIFIER','SEQUENCE_NO'])
    d=train[train['UNIQUE_IDENTIFIER']==i]
    x1=np.max(d['Y1'].to_numpy())
    x2=np.max(d['Y2'].to_numpy())
    arr=df.to_numpy()
    p=np.shape(arr)
    X_train[j,:p[0],:]=arr[:,:]
    Y_train[j,0]=x1
    Y_train[j,1]=x2
    j=j+1
```

```python
learning_rate = 0.01    
training_steps = 10000
batch_size = 10      
display_freq = 1000 
learning_rate = 0.001 
epochs = 10           
batch_size = 100     
display_freq = 100  
```

```python
num_hidden_units = 10 
```

```python
def weight_variable(shape):

    initer = tf.truncated_normal_initializer(stddev=0.01)
    return tf.get_variable('W',dtype=tf.float32,shape=shape,initializer=initer)

def bias_variable(shape):
    
    initial = tf.constant(0., shape=shape, dtype=tf.float32)
    return tf.get_variable('b',dtype=tf.float32,initializer=initial)

def next_batch(x, y, seq_len, batch_size):
    N = x.shape[0]
    batch_indeces = np.random.permutation(N)[:batch_size]
    x_batch = x[batch_indeces]
    y_batch = y[batch_indeces]
    seq_len_batch = seq_len[batch_indeces]
    return x_batch, y_batch, seq_len_batch

def RNN(x, weights, biases, n_hidden, seq_max_len, seq_len):
    
    cell = tf.nn.rnn_cell.BasicRNNCell(n_hidden)
    outputs, states = tf.nn.dynamic_rnn(cell, x, sequence_length=seq_len, dtype=tf.float32)
    batch_size = tf.shape(outputs)[0]
    index = tf.range(0, batch_size) * seq_max_len + (seq_len - 1)
    outputs = tf.gather(tf.reshape(outputs, [-1, n_hidden]), index)
    out = tf.matmul(outputs, weights) + biases
    return out
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()
input_dim=38
x = tf.placeholder(tf.float32, [None, seq_max_length, input_dim])
seqLen = tf.placeholder(tf.int32, [None])
y = tf.placeholder(tf.float32, [None, 2])
```

```python
out_dim=2
with tf.variable_scope("W", reuse=tf.AUTO_REUSE):
    W = weight_variable(shape=[num_hidden_units, out_dim])

with tf.variable_scope("b", reuse=tf.AUTO_REUSE):
    b = bias_variable(shape=[out_dim])

pred_out = RNN(x, W, b, num_hidden_units, seq_max_length, seqLen)
```

```python
cost = tf.reduce_mean(tf.square(pred_out - y))
train_op = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)
```

```python
init = tf.global_variables_initializer()
```

```python
index1=pd.unique(test['UNIQUE_IDENTIFIER'])
count_test=int(len(index1))
X_test=np.zeros((count_test, seq_max_length, 38))
seq_length_test=np.zeros((count_test))
j=0
for i in index1:
    df=test[test['UNIQUE_IDENTIFIER']==i]
    seq_length_test[j]=np.max(df['SEQUENCE_NO'].to_numpy())
    df=df.drop(columns=['UNIQUE_IDENTIFIER','SEQUENCE_NO'])
    d=test[test['UNIQUE_IDENTIFIER']==i]
    arr=df.to_numpy()
    p=np.shape(arr)
    X_test[j,:p[0],:]=arr[:,:]
    j=j+1
```

```python
training_steps=100000
with tf.Session() as sess:
    sess.run(init)
    print('----------Training---------')
    for i in range(training_steps):
        x_batch, y_batch, seq_len_batch = next_batch(X_train, Y_train, seq_length_train, batch_size)
        _, mse = sess.run([train_op, cost], feed_dict={x: x_batch, y: y_batch, seqLen: seq_len_batch})
        if i % display_freq == 0:
            print('Step {0:<6}, MSE={1:.4f}'.format(i, mse))
    y_pred = sess.run(pred_out, feed_dict={x: X_test, seqLen: seq_length_test})
```
